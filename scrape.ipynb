### Rename

```python
with open('../constitution_scraping/names.txt','r') as inFile:
    names=inFile.read().split('\r\n')
```

```python
names.pop(-1)
```

```python
def getCountry(f):
    for name in names:
        if re.search(name.lower(),f.lower()):
            return name
    return None
```

```python
def move(f,country):
    if not os.path.exists('%s.pdf' % country):
        os.rename('%s' % f,'files/%s.pdf' % country)
    else:
        os.rename('%s' % f,'files/%s_2.pdf' % country)
```

```python
for f in glob.glob('files/*'):
    country=getCountry(f)
    print country,': ',f
    
    if country:
        move(f,country)
```
